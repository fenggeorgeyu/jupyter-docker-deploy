```python
import pylab
ax=[0., 0.5, 1.0, 1.5, 2.0, 2.5, 3.0]
ay=[0.0, 0.25, 1.0, 2.25, 4.0, 6.25, 9.0]
pylab.plot(ax, ay)
pylab.show()
```

```python
import random
ax, ay = [], []
for i in range(100):
    ax.append(random.random())
    ay.append(random.random())
pylab.scatter(ax,ay)
pylab.show()
```

```python
import pylab
pylab.rc('text',usetex=True)#enable latex
x=pylab.linspace(-10,10,1001)
for n in range(1,5):
    y = x**n * pylab.sin(x)
    y /= max(y)
    pylab.plot(x,y,label=r'$x^{}\sin x$'.format(n))
pylab.legend(loc='lower center')
pylab.show()
# pylab.savefig('sinnx.pdf',format='pdf')# save to pdf file
# pylab.savefig('sinnx.svg',format='svg',transparent=True)# save to svg file
```
